```python
# Easier way to write
#mpg = mpg.assign(mileage_difference = mpg.hwy - mpg.cty,
#                average_mileage = (mpg.cty + mpg.hwy) /2)

mpg['mileage_difference'] = mpg.hwy - mpg.cty
mpg.head()
```

```python
mpg['average_mileage'] = (mpg.cty + mpg.hwy) / 2

mpg.head()
```

```python
mpg['is_automatic'] = mpg.trans.str.startswith('a')
mpg.head()
```

```python

mpg.groupby(['manufacturer', 'average_mileage']).average_mileage.agg('max').nlargest(1)
```

```python
mpg.groupby('manufacturer').average_mileage.max().nlargest(1)
```

```python
a_mileage = mpg.trans.str.startswith('a')

m_mileage = mpg.trans.str.startswith('m')

a_mpg = mpg[a_mileage].average_mileage.agg('mean')

m_mpg = mpg[m_mileage].average_mileage.agg('mean')

a_mpg > m_mpg

```
