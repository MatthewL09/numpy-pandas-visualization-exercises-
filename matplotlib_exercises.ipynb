```python
import matplotlib.pyplot as plt

import math as m
import numpy as np
import pandas as pd

```

```python
# number 1

numbers = range(-100,100)

def f(x):
    return x ** 2 - x + 2

y = [f(x) for x in numbers]

plt.figure(figsize = (7,5))

plt.grid(True, ls = '--')
plt.plot(numbers, y, color = 'green')
plt.annotate('origin', xy = (0,0))
plt.show()
```

```python
# number 2.1


numbers = range(1, 50)

def s(x):
    square = m.sqrt(x)
    return square

y = [s(x) for x in numbers]

plt.plot(numbers, y)
plt.title('Square root canvas')
plt.show()
```

```python
# number 2.2

numbers = range(-100, 100)

def t(x):
    return x ** 3

y = [t(x) for x in numbers]

plt.plot(numbers, y)
plt.title('To the power of 3 canvas')
plt.show()
```

```python
# number 2.3

numbers = range (1, 50)

def p(x):
    num = 2 ** x
    return num

y = [p(x) for x in numbers]

plt.plot(numbers, y)
plt.title('The power of X canvas')
```

```python
# number 2.4

numbers = (1,50)

def n(x):
    values = 1 / (x + 1)
    return values

y = [n(x) for x in numbers]

plt.plot(numbers, y)
plt.title('Dividend of X + 1')
```

```python
# number 3
plt.figure(figsize = (16,12))

plt.subplot(221)

numbers = range(1, 50)
y = [s(x) for x in numbers]
y2 = [t(x) for x in numbers]
y3 = [p(x) for x in numbers]
y4 = [n(x) for x in numbers]

plt.grid(True, ls = '--')
plt.plot(numbers, y, color = 'blue')
plt.title('Square root canvas')


plt.subplot(222)

plt.grid(True, ls = '--')
plt.plot(numbers, y2, color = 'green')
plt.title('The power of X canvas')


plt.subplot(223)

plt.grid(True, ls = '--')
plt.plot(numbers, y3, color = 'orange')
plt.title('To the power of 3 canvas')

plt.subplot(224)

plt.grid(True, ls = '--')
plt.plot(numbers, y4, color = 'red')
plt.title('Dividend of X + 1')



plt.tight_layout()
```

```python
# number 4

numbers = range(0,10)
y = [s(x) for x in numbers]
y2 = [t(x) for x in numbers]
y3 = [p(x) for x in numbers]
y4 = [n(x) for x in numbers]

plt.figure(figsize = (10,8))

plt.plot(numbers, y, color ="blue", label = '$y = sqrt(x)$')
plt.plot(numbers, y2, color ="green", label = '$y = x^3$')
plt.plot(numbers, y3, color = "orange", label = '$y= 2^x$')
plt.plot(numbers, y4, color ="red", label = '$y = 1 /(x+1)$')

plt.xlim(0,5)
plt.ylim(0, 15)
plt.grid(True, ls ='--')
plt.title('What does y = ?')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()
```
